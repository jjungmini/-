```python
# 필요한 라이브러리 임포트
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import ElasticNetCV
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import numpy as np
import matplotlib.pyplot as plt
import h2o
from h2o.automl import H2OAutoML
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
from sklearn.model_selection import train_test_split

pd.set_option('display.max_columns', 100)
pd.set_option('display.max_rows', 100)

import warnings
warnings.filterwarnings('ignore')
```

```python
# 기상 데이터 불러오기
weather_path = '/content/AWS 삼척 시간단위 기상데이터'
weather_list = []

for file in os.listdir(weather_path):
    weather_list.append(pd.read_csv(weather_path + '/' + file, encoding='cp949'))

weather = pd.concat(weather_list)
weather.index = pd.to_datetime(weather['일시'])
weather = weather.drop(columns=['지점', '지점명', '풍향(deg)', '일시', '현지기압(hPa)', '해면기압(hPa)'])
weather = weather.sort_index()
weather
```

```python
# 태양광 발전량 데이터 불러오기
solar = pd.read_csv('태양광 발전량 데이터.csv')
solar.index = pd.to_datetime(solar['일시'])
solar = solar.drop(columns=['일시'])
solar
```

```python
# 추가적으로 수집된 기상 데이터 불러오기
my_data = pd.read_excel('자체 수집 데이터.xlsx')

# 날짜를 datetime 타입으로 변환
my_data['일시'] = pd.to_datetime(my_data['일시'], format='%Y-%m-%d-%H')
```

```python
# 데이터 병합
from functools import reduce
df = reduce(lambda x,y: pd.merge(x,y, on='일시', how='outer'), [solar, weather, my_data])
df
```

```python
# 발전소 데이터의 결측치를 0으로 채우기
df[['AC_출력_1', 'AC_출력_2']] = df[['AC_출력_1', 'AC_출력_2']].fillna(0)

# columns 이름 변경
df.rename(columns={'기온(°C)': 'temperature', '풍속(m/s)': 'windspeed', '강수량(mm)':'Precipitation', '습도(%)':'Humidity'}, inplace=True)

# 나머지 열 선형보간법으로 결측치 채우기
for i in df.columns[3:7]: # ['temperature', 'windspeed', 'Precipitation', 'Humidity']
    df[i] = df[i].interpolate(method='linear', limit_direction='forward')

#결측치 확인
df.info()
```

```python
# scaled data frame 생성
df_scaled = df.copy()

# solarpower 변수를 위한 별도의 스케일러 정의
solar_scaler = MinMaxScaler()
df_scaled['solarpower'] = solar_scaler.fit_transform(df_scaled[['solarpower']])

# 특성만 정규화하기 위해 Scaler 정의
scaler = MinMaxScaler()

# 특성들 간의 상관관계 계산을 위해 타겟 변수 제외
correlation_matrix = df_scaled.drop(columns=['일시', 'solarpower', 'AC_출력_1', 'AC_출력_2']).corr()

# Solarpower 제외하고 정규화
features = df_scaled.drop(columns=['일시', 'solarpower', 'AC_출력_1', 'AC_출력_2'])
scaled_features = scaler.fit_transform(features)
df_scaled[features.columns] = scaled_features

# 상관관계 히트맵 그리기
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python
# 훈련 데이터와 테스트 데이터로 분리
train = df_scaled[df_scaled['일시'] < '2023-01-01']
test = df_scaled[(df_scaled['일시'] >= '2023-01-01') & (df_scaled['일시'] < '2023-04-01')]
```

```python
# 훈련 데이터를 이용하여 특성 선택을 위한 ElasticNetCV 모델을 생성합니다
enet_train = ElasticNetCV(
    alphas=[0.001, 0.01, 0.1, 1, 10],        # 후보 alpha 값 리스트
    l1_ratio=[0.1, 0.3, 0.5, 0.7, 0.9],     # 후보 l1_ratio 값 리스트
    cv=5,                                   # 5-fold cross-validation 사용
    n_alphas=100,                           # 100개의 alpha 값 생성
    max_iter=1000,                          # 최대 반복 횟수
    tol=1e-4,                               # 수렴 기준
    verbose=False,                          # 로그 출력 비활성화
    precompute='auto',                      # 자동으로 사전 계산된 Gram 행렬 결정
    selection='random',                      # 랜덤하게 특성 선택
    random_state=42
)
enet_train.fit(train.drop(columns=['일시', 'solarpower', 'AC_출력_1', 'AC_출력_2']), train['solarpower'])
# alpha 값을 출력합니다.
print('Best alpha: ', enet_train.alpha_)

# 선택된 특성만 남김
selected_features_train = train.drop(columns=['일시', 'solarpower', 'AC_출력_1', 'AC_출력_2']).columns[enet_train.coef_ != 0]

# 선택된 특성 출력
print("Selected features: ", selected_features_train.tolist())

# 선택된 특성으로 데이터프레임 생성
df_part = df[['일시', 'AC_출력_1', 'AC_출력_2'] + selected_features_train.tolist()]
```

```python
# ElasticNetCV의 결과를 출력합니다.
# 특성 중요도를 시각화합니다.
plt.figure(figsize=(10, 6))
plt.barh(range(len(enet_train.coef_)), enet_train.coef_)
plt.xlabel('Feature Importance')
plt.ylabel('Features')
plt.yticks(range(len(enet_train.coef_)), enet_train.feature_names_in_)
plt.title('Feature importance by ElasticNetCV (Train Data)')
plt.show()
```

```python
# df_part(변수 선택 후 데이터)에서 훈련 데이터와 테스트 데이터로 분리
train = df_part[df_part['일시'] < '2023-01-01']
test = df_part[(df_part['일시'] >= '2023-01-01') & (df_part['일시'] < '2023-04-01')]

# 예측 결과를 저장할 딕셔너리 선언
y_pred_tr_dict = {}
y_pred_te_dict = {}

# MAPE 계산 함수 정의
def mean_absolute_percentage_error(y_true, y_pred):
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100

# 반복문 사용 문자 정의
k = 0

# 모델 학습
for i in ['AC_출력_1', 'AC_출력_2']:

  # y값을 위한 별도의 스케일러 정의
  y_scaler = MinMaxScaler()
  train[i] = y_scaler.fit_transform(train[[i]])

  # x값에 대한 별도의 스케일러 정의
  x_scaler = MinMaxScaler()
  features = train.drop(columns=['일시', 'AC_출력_1', 'AC_출력_2'])
  scaled_features = x_scaler.fit_transform(features)
  train[features.columns] = scaled_features

  # Test 데이터 스케일러 적용
  test[i] = y_scaler.transform(test[[i]])
  features = test.drop(columns=['일시', 'AC_출력_1', 'AC_출력_2'])
  scaled_features = x_scaler.transform(features)
  test[features.columns] = scaled_features

  # H2O 라이브러리 초기화
  h2o.init()

  # H2O 데이터프레임으로 변환
  option = ['AC_출력_2', 'AC_출력_1']
  h2o_train = h2o.H2OFrame(train.drop(columns=['일시', option[k]], axis=1))

  # H20 데이터 프레임 저장
  file_path = f"/content/입력데이터/{i}_입력데이터_train.csv"
  h2o.export_file(h2o_train, path=file_path)

  # h2o 데이터 프레임 train)

  # Date 열을 제외한 특성과 타겟 변수를 정의
  x = h2o_train.columns
  x.remove(i)
  print(x)
  y = i

  # AutoML 설정 및 모델 훈련 (모든 파라미터를 원하는 값으로 변경하여 설정)
  aml = H2OAutoML(
     max_runtime_secs=2400,     # AutoML 실행에 허용할 최대 시간 (초)
    # max_models=20,             # 생성할 최대 모델 수
    stopping_metric='RMSE',    # 조기 종료를 위한 평가 지표
    stopping_tolerance=0.01,  # 조기 종료를 위한 허용 오차
    stopping_rounds=4,         # 조기 종료를 위한 반복 횟수
    nfolds=5,                  # 교차 검증(fold)의 개수
    seed=42,                   # 랜덤 시드
    # exclude_algos = ["StackedEnsemble"],

    include_algos=["StackedEnsemble", 'XGBoost', 'GBM'],  # 사용할 알고리즘들의 리스트
    keep_cross_validation_predictions=True,  # 교차 검증 예측 결과 보존 여부
    keep_cross_validation_models=False,      # 교차 검증 모델 보존 여부
    keep_cross_validation_fold_assignment=False,  # 교차 검증 폴드 할당 보존 여부
    sort_metric='RMSE',        # 성능 측정 기준
    export_checkpoints_dir=None  # 모델 체크포인트를 내보낼 디렉토리
  )

  aml.train(x=x, y=y, training_frame=h2o_train) #모델 학습

  # AutoML 결과 확인
  # AutoML 결과 확인 (모든 모델 출력)
  lb = aml.leaderboard
  print(lb)

  # Leaderboard 데이터를 Pandas DataFrame으로 변환
  leaderboard_df = lb.as_data_frame()

  # 모든 모델의 성능 순위를 변수로 저장
  for idx, row in leaderboard_df.iterrows():
    model_id = row['model_id']
    rank = idx + 1
    exec(f"model_rank_{rank} = h2o.get_model('{model_id}')")

  # 모든 모델의 성능 순위 출력
  print("\n<< 모든 모델의 성능 순위 >>")
  for idx, row in leaderboard_df.iterrows():
    model_id = row['model_id']
    rank = idx + 1
    print(f"Model: {model_id}, Rank: {rank}")

  # 가장 성능이 좋은 모델 선택
  best_model = aml.leader

  # 모델 세부 정보 출력
  print("Best model is: ", best_model)

  # 훈련된 모델 저장
  model_path=f"/content/모델/{i}_model"
  my_local_model = h2o.download_model(best_model, path=model_path)

  # 원하는 rank의 모델을 사용하여 train 데이터 예측 수행
  predictions = best_model.predict(h2o_train) #model_rank_n
  predictions = predictions.as_data_frame()

  # 예측된 값을 원래 스케일로 역변환
  predictions = y_scaler.inverse_transform(predictions)

  # Train 데이터 예측 결과 딕셔너리 저장
  y_pred_tr_dict[i] = predictions

  # train 데이터셋의 실제 타겟 변수 값을 가져옴
  actual_values = train[i].values.reshape(-1, 1)
  actual_values = y_scaler.inverse_transform(actual_values)

  # 평가 지표 계산
  mse = mean_squared_error(actual_values, predictions)
  mae = mean_absolute_error(actual_values, predictions)
  r2 = r2_score(actual_values, predictions)
  mape = mean_absolute_percentage_error(actual_values, predictions)
  rmse = np.sqrt(mse)

  # 평가 지표 출력
  print("평균 제곱 오차 (MSE):", mse)
  print("평균 절대 오차 (MAE):", mae)
  print("결정 계수 (R2):", r2)
  print("평균 절대 백분율 오차 (MAPE):", mape)
  print("평균 제곱근 오차 (RMSE):", rmse)

  # 실제 값과 예측값 시각화
  plt.figure(figsize=(20, 6))
  plt.plot(train['일시'], actual_values, label='Actual', marker='o')
  plt.plot(train['일시'], predictions, label='Forecasted', marker='o')
  plt.xlabel('일시')
  plt.ylabel('Power Generation')
  plt.title(f'Actual vs. Forecasted(train_{i})')
  plt.legend()
  plt.show()


  # 테스트 데이터를 H2O 데이터프레임으로 변환
  h2o_test = h2o.H2OFrame(test.drop(columns=['일시', option[k]], axis=1))
  k= k+1

  # H20 데이터 프레임 저장(test)
  file_path = f"/content/입력데이터/{i}_입력데이터_test.csv"
  h2o.export_file(h2o_test, path = file_path)

  # 원하는 rank의 모델을 사용하여 예측 수행
  predictions = best_model.predict(h2o_test) #model_rank_n
  predictions = predictions.as_data_frame()

  # 예측된 값을 원래 스케일로 역변환
  predictions = y_scaler.inverse_transform(predictions)

  # Test 데이터 예측결과를 딕셔너리에 저장
  y_pred_te_dict[i] = predictions

  # 테스트 데이터셋의 실제 타겟 변수 값을 가져옴
  actual_values = test[i].values.reshape(-1, 1)
  actual_values = y_scaler.inverse_transform(actual_values)

  # 평가 지표 계산
  mse = mean_squared_error(actual_values, predictions)
  mae = mean_absolute_error(actual_values, predictions)
  r2 = r2_score(actual_values, predictions)
  mape = mean_absolute_percentage_error(actual_values, predictions)
  rmse = np.sqrt(mse)

  # 평가 지표 출력
  print("평균 제곱 오차 (MSE):", mse)
  print("평균 절대 오차 (MAE):", mae)
  print("결정 계수 (R2):", r2)
  print("평균 절대 백분율 오차 (MAPE):", mape)
  print("평균 제곱근 오차 (RMSE):", rmse)

  # 실제 값과 예측값 시각화
  plt.figure(figsize=(20, 6))
  plt.plot(test['일시'], actual_values, label='Actual', marker='o')
  plt.plot(test['일시'], predictions, label='Forecasted', marker='o')
  plt.xlabel('일시')
  plt.ylabel('Power Generation')
  plt.title(f'Actual vs. Forecasted(test_{i})')
  plt.legend()
  plt.show()
```

```python
# NMAE 평가지표 정의 함수
def NMAE_FUNC (true, pred, S):

    # 판다스 데이터프레임 기준
    # true : 실제발전량
    # pred : 예측발전량
    # S : 발전용량(kW)

    true_values = true.loc[true.values > S * 0.1, :]
    pred_values = pred.loc[true_values.index, :]

    AE = sum(abs(true_values.values - pred_values.values))
    NAE = AE / S
    NMAE = 100 * (NAE / len(true_values))

    return NMAE[0]
```

```python
# 딕셔너리의 값에 대해 리스트 평탄화
data_flat_tr = {key: [item[0] for item in value] for key, value in y_pred_tr_dict.items()}
# 평탄화된 딕셔너리를 데이터프레임으로 변환
pred_tr = pd.DataFrame(data_flat_tr)

# 딕셔너리의 값에 대해 리스트 평탄화
data_flat_te = {key: [item[0] for item in value] for key, value in y_pred_te_dict.items()}
# 평탄화된 딕셔너리를 데이터프레임으로 변환
pred_te = pd.DataFrame(data_flat_te)
```

```python
pred_tr.columns = ['예측_발전량_1', '예측_발전량_2']
pred_te.columns = ['예측_발전량_1', '예측_발전량_2']
pred_tr['예측총발전량'] = pred_tr.sum(axis=1)
pred_te['예측총발전량'] = pred_te.sum(axis=1)

test_date = '2022-12-31 23:00:00'
true_tr = df_part[['AC_출력_1', 'AC_출력_2']].loc[df_part["일시"] <= test_date, :].reset_index(drop=True)
true_te = df_part[['AC_출력_1', 'AC_출력_2']].loc[df_part["일시"] > test_date, :].reset_index(drop=True)

true_tr['총발전량'] = true_tr.sum(axis=1)
true_te['총발전량'] = true_te.sum(axis=1)


```

```python
S_total = 100 # 총 설비용량 : 100kW

# Train 예측총발전량에 대한 NMAE(%)
tr_NMAE = NMAE_FUNC(true=true_tr[['총발전량']], pred=pred_tr[['예측총발전량']], S=S_total)
print('Train NMAE(%) :', tr_NMAE, '%')

# Test 예측총발전량에 대한 NMAE(%)
te_NMAE = NMAE_FUNC(true=true_te[['총발전량']], pred=pred_te[['예측총발전량']], S=S_total)
print('Test NMAE(%) :', te_NMAE, '%')
```

```python
# 평가 지표 출력
print("평균 제곱 오차 (MSE):", mse)
print("평균 절대 오차 (MAE):", mae)
print("결정 계수 (R2):", r2)
print("평균 절대 백분율 오차 (MAPE):", mape)
print("평균 제곱근 오차 (RMSE):", rmse)

# 실제 값과 예측값 시각화
plt.figure(figsize=(20, 6))
plt.plot(test['일시'], true_te[['총발전량']], label='Actual', marker='o')
plt.plot(test['일시'], pred_te[['예측총발전량']], label='Forecasted', marker='o')
plt.xlabel('일시')
plt.ylabel('Power Generation')
plt.title('Actual vs. Forecasted(total)')
plt.legend()
plt.show()
```

```python
# Train, Test 결과 저장
pred_tr.to_csv('Train 예측 결과.csv', encoding='utf-8-sig')
pred_te.to_csv('Test 예측 결과.csv', encoding='utf-8-sig')

# Train, Test NMAE(%) 결과 저장
pd.DataFrame([tr_NMAE, te_NMAE], index=['Train_NMAE(%)', 'Test_NMAE(%)']).T.to_csv('NMAE_결과.csv', index=False, encoding='utf-8-sig')
```
